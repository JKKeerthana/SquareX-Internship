# **Part 1: SaaS Discovery & Overview Dashboard**

```python
saas_category_filters = """
(
  (
    url_category LIKE '%Productivity%' OR
    url_category LIKE '%Business/Economy%' OR
    url_category LIKE '%File Sharing/Personal Storage%' OR
    url_category LIKE '%Email%' OR
    url_category LIKE '%Generative AI%' OR
    url_category LIKE '%AI/ML Applications%' OR
    url_category LIKE '%IT%'
    )
    AND NOT (
      url_category LIKE '%Social Networking%' OR
      url_category LIKE '%Entertainment%' OR
      url_category LIKE '%News and Media%' OR
      url_category LIKE '%Games%' OR
      url_category LIKE '%Personal Sites%' OR
      url_category LIKE '%Shopping%' OR
      url_category LIKE '%Media Sharing%'
      )
)
"""

saas_idp_filters = """
identity_provider IN ('Google', 'Microsoft', 'GitHub', 'Apple', 'Okta')
"""

saas_filter = f"(({saas_category_filters}) OR ({saas_idp_filters}))"
```

```python
# Top SaaS apps used
query_top_saas = f"""
SELECT domain, COUNT(*) AS usage_count
FROM events
WHERE {saas_filter} AND domain IS NOT NULL AND domain != ''
GROUP BY domain
ORDER BY usage_count DESC
LIMIT 20;
"""
top_saas = pd.read_sql_query(query_top_saas, con)
top_saas
```

```python
# SaaS apps with HIGH severity
query_high_risk = f"""
SELECT domain, COUNT(*) AS high_risk_events
FROM events
WHERE {saas_filter} AND severity IN ('HIGH', 'CRITICAL') AND domain IS NOT NULL
GROUP BY domain
ORDER BY high_risk_events DESC
LIMIT 10;
"""
high_risk = pd.read_sql_query(query_high_risk, con)
high_risk
```

```python
# SaaS apps blocked
query_blocked = f"""
SELECT domain, COUNT(*) AS blocked_count
FROM events
WHERE {saas_filter} AND effect = 'Block' AND domain IS NOT NULL
GROUP BY domain
ORDER BY blocked_count DESC
LIMIT 10;
"""
blocked = pd.read_sql_query(query_blocked, con)
blocked
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(12,5))
plt.bar(top_saas['domain'], top_saas['usage_count'], color='steelblue')
plt.xticks(rotation=90)
plt.title("Top 20 SaaS Domain Usage by Employees")
plt.ylabel("Number of Events")
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(10,4))
plt.bar(high_risk['domain'], high_risk['high_risk_events'], color='darkred')
plt.xticks(rotation=90)
plt.title("Top 10 Domains With HIGH Severity Events")
plt.ylabel("Number of HIGH Severity Events")
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(10,4))
plt.bar(blocked['domain'], blocked['blocked_count'], color='orange')
plt.xticks(rotation=90)
plt.title("Top 10 Domains With Blocked Actions")
plt.ylabel("Number of Blocked Events")
plt.tight_layout()
plt.show()
```

***SaaS Discovery & Overview***

**Method:**

*  We defined SaaS apps by selecting domains from business-relevant categories (Productivity, IT, File Sharing, etc.) and known identity providers (Google, Microsoft, Okta, GitHub, Apple), while excluding Social, Entertainment, Shopping, etc.

* Using SQL, counted usage events, high-severity alerts, and blocked actions to highlight commonly used apps and potential security risks.

**Key** **Insights**:

*  Top SaaS apps used by employees are google.com, chatgpt.com, jetbrains.com, live.com, etc.


* Some apps have many security warnings. For example, ChatGPT.com and Vercel.app raised a lot of “HIGH” risk alerts.

* Some apps often get blocked by security rules. Google.com, ChatGPT.com, and Live.com are blocked the most.

* Should we be worried? Yes, you should be cautious: The top-used SaaS apps are also frequently blocked or flagged for HIGH severity, it indicates possible risky behavior or configuration gaps.

**Limitation:**

* The SaaS app classification with URL Category/Identity Provider may misclassify edge cases or miss some SaaS apps.


# **Part 2: Sensitive Data Leakage on GenAI Apps**

```python
df['sensitive_data_classification'].value_counts()
```

```python
query_sensitive = """
SELECT domain, sensitive_data_classification,
  CASE
    WHEN sensitive_data_classification IN ('email', 'phone', 'address', 'SIN') THEN 'PII'
    WHEN sensitive_data_classification = 'credit card number' THEN 'Financial'
    WHEN sensitive_data_classification = 'python' THEN 'Source Code'
    WHEN input_is_credentials = 1 THEN 'Credentials'
    WHEN clipboard_text LIKE '%sk-%'
        OR clipboard_text LIKE '%api_key%'
        OR clipboard_text LIKE '%token%'
        OR clipboard_text LIKE '%Bearer %' THEN 'Secrets/API Keys'
    ELSE 'Other Sensitive'
  END AS expanded_sensitive_category,
  COUNT(*) AS count
FROM events
WHERE domain IN ('chatgpt.com', 'claude.ai', 'deepseek.com', 'perplexity.ai', 'openai.com',
                 'sider.ai', 'glasp.co', 'otter.ai', 'grammarly.com', 'bard.google.com')
  AND (sensitive_data_classification IS NOT NULL
        OR input_is_credentials = 1
        OR clipboard_text LIKE '%sk-%'
        OR clipboard_text LIKE '%api_key%'
        OR clipboard_text LIKE '%token%'
        OR clipboard_text LIKE '%Bearer %')
GROUP BY domain, sensitive_data_classification, expanded_sensitive_category
ORDER BY count DESC;
"""
sensitive_data = pd.read_sql_query(query_sensitive, con)
sensitive_data
```

```python
import seaborn as sns
# Overall Leakage by Sensitive Data Type
plt.figure(figsize=(10,6))
sns.barplot(x='expanded_sensitive_category', y='count', data=sensitive_data, estimator=sum, errorbar=None)
plt.xticks(rotation=45)
plt.title("Sensitive Data Leakage by Type")
plt.xlabel("Sensitive Data Type")
plt.ylabel("Number of Events")
plt.show()
```

```python
# Heatmap: Domains vs Data Types
heatmap_df = sensitive_data.groupby(['expanded_sensitive_category', 'domain'])['count'].sum().unstack(fill_value=0)
sns.heatmap(heatmap_df, annot=True, cmap='Reds')
plt.title("GenAI Domains vs Types of Sensitive Data Leaked")
plt.xlabel("GenAI Domain")
plt.ylabel("Sensitive Data Type")
plt.tight_layout()
plt.show()
```

***Sensitive Data Leakage on GenAI Apps***

**Method**:

* Focused on key GenAI domains and classified sensitive data into categories like PII, Financial, Source Code, Credentials, and Secrets using existing fields and keyword patterns. SQL queries aggregated sensitive data events by domain and type, and visualizations reveal data leakage risks clearly.

**Key** **Insights**:

* Sensitive information (like emails, phone numbers, credit card numbers, addresses, and even Python code) has been sent to GenAI apps, especially ChatGPT.com.

* Most leaks involve personal info (PII), some financial data, and source code.

* ChatGPT.com has the highest amount of sensitive data sent by far.

* Sharing confidential or personal information with AI apps can put the company at risk. This can leak customer data, secrets, or private details.

**Limitation:**

* Analysis focused on known GenAI domains other, lesser-known AI tools might not be captured.
